```python
import numpy as np
import matplotlib.pyplot as plt
from perceptron import Perceptron

plt.style.use("default")
```

```python
x_train = np.array([[3, 3], [4, 3], [1, 1]])
y_train = np.array([1, 1, -1])
```

```python
mode = Perceptron()
mode.fit(x_train=x_train, y_train=y_train)
```

```python
positive_samples = x_train[y_train == 1]
negative_samples = x_train[y_train == -1]
x1 = np.linspace(0, 7, 7)
x2 = -x1 - mode._b
plt.scatter(positive_samples[:, 0], positive_samples[:, 1], color="green")
plt.scatter(negative_samples[:, 0], negative_samples[:, 1], color="red")
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.plot(x1, x2)
```

```python
mode.predict(np.array([[2, 3], [4, 5], [5, 5], [2, 2], [1, 1]]))
```
